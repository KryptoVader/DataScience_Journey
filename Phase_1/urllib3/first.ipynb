```python
import urllib3 as url
import requests
```

# Pool Manager
In urllib3 we have to create the poolmanager to handle multiple requests along with thread safety

```python
http = url.PoolManager()
```

```python
r = http.request('GET','https://httpbin.org/robots.txt')
print(r.data)
```

```python
r = http.request('POST','https://httpbin.org/post', fields={'hi':'hello'})
print(r.data)
```

```python
r.json()
```

```python
r.headers
```

```python
import io
```

```python
resp = http.request("GET", "https://example.com", preload_content=False)
resp.auto_close = False

for line in io.TextIOWrapper(resp):
    print(line)
```

# SENDING FILE USING URLLIB3

```python
with open('example.txt', 'r') as f:
    file_data = f.read()
    
conn = http.request('POST', 'https://httpbin.org/post', fields={'filefield' : ('example.txt', file_data)})
print(conn.json()['files'])
```

```python
conn.json()
```
