## Seperation of Category_code

```python
ecom.head()
```

```python
ecom[['category', 'type']] = ecom['category_code'].str.extract(r'^([^\.]+)\.?(.*)$')
ecom
```

```python
ecom['category'] = ecom['category'].fillna('unknown')
ecom['type'] = ecom['type'].fillna('unknown')
ecom['brand'] = ecom['brand'].fillna('unknown')
ecom
```

```python
ecom = ecom[ecom['user_session'].notna()]
```

```python
ecom = ecom.drop(columns=['category_code'])
```

```python
ecom.isna().sum()
```

```python
ecom.head()
```

```python
import os
import pyarrow as pa
import pyarrow.parquet as pq
import gc

chunk_size = 100_000  # No downcasting, so keep this small
n_chunks = len(ecom) // chunk_size + 1
save_path = r"C:/Datasets/Parquets/ecom_chunks"

os.makedirs(save_path, exist_ok=True)

for i in range(n_chunks):
    chunk = ecom.iloc[i * chunk_size : (i + 1) * chunk_size]

    # Optional: drop unused object references, clear cache
    gc.collect()

    table = pa.Table.from_pandas(chunk, preserve_index=False)

    pq.write_table(
        table,
        f"{save_path}/ecom_part{i}.parquet",
        row_group_size=50_000  # Helps Arrow avoid large memory blocks
    )

    # Clear memory for next loop
    del chunk
    del table
    gc.collect()
```

```python
import pyarrow.parquet as pq
import pyarrow as pa
import glob
import os

# Paths
chunks_path = "C:/Datasets/Parquets/ecom_chunks/ecom_part*.parquet"
output_file = "C:/Datasets/Parquets/ecom.parquet"

# Get list of chunk files
all_files = sorted(glob.glob(chunks_path))

# Open writer only once, append each chunk
writer = None

for i, file in enumerate(all_files):
    table = pq.read_table(file)

    if writer is None:
        # First chunk — create writer
        writer = pq.ParquetWriter(output_file, table.schema)
    
    # Write this chunk
    writer.write_table(table)

# Finalize the writer
if writer:
    writer.close()

print("✅ Merged all Parquet chunks into one file successfully.")
```
